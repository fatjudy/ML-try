## Regular Expressions - RegEx

**A regular expression (or RE) specifies a set of strings that matches it.**

In this section we will be learning about the python module "re". The functions in this module let you check if a particular string matches a given regular expression.

RegEx online tool: https://www.regexpal.com/

Python "re" documentation: https://docs.python.org/3/library/re.html




### A shortlist of RegEx syntax

    . (Period)
A period matches any single character (except newline '\n').

    [] (Square brackets)
Square brackets specifies a set of characters you wish to match. For example, [xyz] will match if the string is any combination of x, y, or z. [^xyz] means any string except x, y, or z. [^a-z] means any non-lowercase alphabet.

    ^ (Caret)
A caret (by itself, not within square brackets) is used to check if a string starts with a certain character (or set of characters if you use square brackets).

    $ (Dollar)
The dollar symbol is used to check if a string ends with a certain character. For example, x$ checks if the string ends with x.

    * (Asterisk)
Matches zero or more occurences of the pattern to the left of it. For example, ye*s matches with ys, yes, yees, yeees.

    + (Plus)
Matches one or more occurences of the pattern to the left of it.

    ? (Question mark)
Matches zero or one occurence of the pattern to the left of it.

    {m,n} (Curly braces)
Matches at least m, at most n repetitions of the pattern to the left of it.

    x|y (Or)
Matches for x or y.

    () (Round braces)
Used to group set of characters (as with typical use in math).





### A shortlist of RegEx methods

## Word2Vec
Typically, you would convert each unique word in your dictionary of words into its own unique number. For example the sentence "I love myself" could become [0, 1, 2]. However, these numbers do not describe the relation between words. The word "I" clearly refers to the same entity as "myself", however, their numbers (0 and 2) are far apart compared to the other word "love" (1).

It is possible to convert each word into a *vector* of numbers. This provides much more flexibility in "describing" a word by its numbers. The aim is to assign numbers to words such that words with similar meanings have numbers that are closer to each other. This makes training a subsequent neural network easier because words have been grouped by semantic meaning.

Word2Vec helps us do all of that!

Word2Vec works by comparing the context of a given word (words surrounding the given word in a sentence) with the context of other words. Words that carry the same meaning tend to appear in similar context and hence will be assigned similar vectors.

There are two models commonly used in Word2Vec: Continuous Bag of Words (CBOW) and Skip-gram. Both methods involve the training of a shallow neural network (single hidden layer) that learns the vector representation of each word. In CBOW, the network is provided the context of a word and trained to predict that word. In Skip-gram, the opposite is holds true - the network is provided the word and is trained to predict which words are likely to surround it. In both cases, what we really want after the training process is the weights of the single hidden layer. These weights represent the vectors learned for each word.

## Formatting our data

Before we start construction of the neural network, we first need to ensure that our data is in a form that is suitable for ingestion by the network. We need to modify our data to satisfy the following criteria:

1.   Every word (token) in our entire dataset (both train and test sets) is translated into a unique number. Recall that neural networks only work with numerical input.
2.   Every input (review text) to the network is of the same length. This is because a recurrent neural network, once trained, is of a constant size with fixed input dimensions. Any input that is larger than this fixed size can be truncated, although that would imply the loss of data. Any input smaller than this fixed size can be padded with zeroes.

The code block below implements these two steps. As you read through the code, watch out for any "TODO" labels. A "TODO" indicates that the code requires editing or filling in before it can run correctly. Make those changes as you deem necessary.

```python
# pip install tensorflow_datasets
```

```python
# Import the necessary packages
import tensorflow as tf
import tensorflow_datasets as tfds
import json

# TODO: Load your processed dataset
dataset = json.load(open('NLP Dataset.json','r'))

# Initilise a TensorFlow tokeniser that automatically tokensises our sentences for us.
tokenizer = tfds.features.text.Tokenizer()

# Create a Python Set to identify and store all unique words in our dataset.
# Any repeated words will be discarded.
vocabulary_set = set() # A Python set() is an unordered collection of unique elements
for review in dataset:
  # Tokenise each sentence into individual words
  tokens = tokenizer.tokenize(review['text'])
  # Add each word into the Set. Repeated words will be discarded
  vocabulary_set.update(tokens)

# vocabulary_set now contains all the unique words in your dataset.
# TODO: Inspect your vocabulary set. How many unique words are there? Are the words in a format that you expect them to be?





# The TokenTextEncoder in tensorflow_datasets (tfds) assigns a unique positive integer to each word in your vocabulary set.
encoder = tfds.features.text.TokenTextEncoder(vocabulary_set)

# TODO: Test the functionality of the encoder by using the encode() method to encode a string of your choice.
# Then use the decode() method to decode an integer into its corresponding word.

encoded_example = encoder.encode('sarcastic')
print('Encoded example: ', encoded_example)

decoded_example = encoder.decode([1062])
print('Decoded example: ', decoded_example)

```

## Building and training a Recurrent Neural Network

Great, now that your data is ready for use, let's design the Recurrent Neural Network.

Building a simple model is relatively easy with TensorFlow Keras. Our model would consist of at least 3 types of layers:


1.   An [Embedding layer](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Embedding) as the first layer of the network. This layer converts each positive integer representing each word in the input into a dense vector of a fixed size. This process is equivalent to the Word2Vec method.
2.   An [LSTM layer](https://www.tensorflow.org/api_docs/python/tf/keras/layers/LSTM) embedded inside of a [Bidirectional layer](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Bidirectional). The Bidirectional layer provides the LSTM layer access to both past and future words in a sentence (both directions), hence facilitating better context and understanding.
3.   [Dense (fully connected) layers](https://keras.io/api/layers/core_layers/dense/) that have a full view of all extracted information to make a prediction.

Have a look at the documentation for these layers linked above, then complete the code for your neural network below. You may add more layers as you wish.



```python
model = tf.keras.Sequential()

# Add an embedding layer. Only one embedding layer is necessary.
# The input dimensions is the vocabulary size + 1 because you have introduced a new word: the empty string '' at index 0.
# The output dimensions is up to you - it represents the length of a vector used to describe each word (similar to Word2Vec).
# Reasonable values for the number of output dimensions may range between 30 and 300 depending on the complexity of your data.
model.add(tf.keras.layers.Embedding(100,30))

# Add an LSTM layer inside of a Bidirectional layer.
# You need to specify the number of units (cells/neurons) that the LSTM layer has.
# A reasonable number of units may range between 50 and 2000 depending on the desired modelling capacity of your neural network.
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(4)))

# Add Dense layers. 
# Again you need to specify the number of neurons in each layer.
# The number of neurons in the hidden layers is a design choice that is up to you.
# This number should be somewhat correlated to the size of the output of the previous layer.
# The final layer should have a specific number of neurons: as many classes as you have.
# You may add more Dense layers as hidden layers as you wish.
model.add(tf.keras.layers.Dense(4, activation='relu'))
model.add(tf.keras.layers.Dense(4))

# Compile the model with a Cross Entropy loss function and perform gradient descent using "Adam".
# We are using "SparseCategoricalCrossentropy" because our labels are in integer form (e.g. 0, 1, 2, 3) 
# and not in one-hot encoded form (e.g. [0,1,0,0], [0,0,0,1], etc).
# This means that each data point in our dataset can belong to one and only one class.
# Also keep note of the learning rate. You may wish to tweak this parameter at a later stage.
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-6),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Finally, train (fit) the model using the training inputs and training labels.
# Testing of the trained model is automatically done at the end of each epoch using the test inputs and test labels.
# Keep note of the batch size and the number of training eopchs. You may wish to tweak these parameters at a later stage.
# 
```
